```python
# library(caret)
#train1 <- createDataPartition(y=data$V58,times=1,p=0.8,list=F)
train_sub=sample(nrow(data),7/10*nrow(data))
train_data <- data[train_sub,]
test_data <- data[-train_sub,]
```

```python
train_data
```

```python
train_data[,57]
```

```python
fit1 <- glm(train_data$V58~.,data=train_data,family="binomial")
summary(fit1)
```

```python
#对测试集进行预测
pred <- as.numeric(predict(fit1,test_data,type="response")>0.5)
pred
```

```python
test_data$V58
```

```python
head(test_data)
```

```python
#将测试集计算所得概率与观测本身取值整合到一起
obs=data.frame(prob=pred,obs=test_data$V58)
```

```python
#输出混淆矩阵
table(test_data$V58,pred,dnn=c("true value","predict value"))
```

```python
install.packages("pROC")
```

```python
#绘制ROC曲线
library('pROC')
logistic_roc <- roc(test_data$V58,pred)
plot(logistic_roc,print.auc=TRUE,auc.polygon=TRUE,grid.col=c("green","red"),max.auc.polygon=TRUE,auc.polygon.col="skyblue",print.thres=TRUE,main='ROC curve')
```

```python
install.packages('randomForest')
```

```python
library(randomForest)
set.seed(666)
fit2 <- randomForest(train_data$V58~.,data=train_data,importance=TRUE)
```

```python
fit2
```

```python
# importance(fit2,type=2)
```

```python
fit2.pred <- as.numeric(predict(fit2,test_data,type="response")>0.5)
fit2.pred
```

```python
fit2.perf <- table(test_data$V58,fit2.pred,dnn=c('Actual','Predicted'))
fit2.perf
```

```python
library('pROC')
logistic_roc <- roc(test_data$V58,fit2.pred)
plot(logistic_roc,print.auc=TRUE,auc.polygon=TRUE,grid.col=c("green","red"),max.auc.polygon=TRUE,auc.polygon.col="skyblue",print.thres=TRUE,main='ROC curve')
```
